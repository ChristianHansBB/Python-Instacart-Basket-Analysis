```python
# Import df_ords_prods_meged
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# create subset

df = ords_prods_merge[:1000000]
```

```python
# Define function price_label
def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# create new empty list 
result_2 = []
```

```python
# The two busiest days are 0 and 1, the two slowest 3 and 4
for value in ords_prods_merge["order_dow"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
# add result_2 as a column
ords_prods_merge['busiest_days'] = result_2
```
