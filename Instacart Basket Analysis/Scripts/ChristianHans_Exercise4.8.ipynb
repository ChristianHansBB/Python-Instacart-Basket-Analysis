## 4.

```python
# Aggregating Data with transform()
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Perform check
ords_prods_merge.head()
```

```python
# Assign loyalty flags
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#checking counts
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Perform check on newest columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(20)
```

## 5.

```python
# Performing Multiple Aggregations
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

New customers spend most while Loyal customers spend least.

## 6.

```python
# Aggregating Data with transform()
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Perform check
ords_prods_merge.head()
```

```python
# Assign average_spend flags
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend'] < 10, 'spender_performance'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend'] >= 10, 'spender_performance'] = 'High spender'
```

```python
#checking counts
ords_prods_merge['spender_performance'].value_counts(dropna = False)
```

```python
# Perform check on newest columns
ords_prods_merge[['user_id', 'average_spend', 'spender_performance']].head(20)
```

## 7.

```python
# Aggregating Data with transform()
ords_prods_merge['frequency_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Perform check
ords_prods_merge.head()
```

```python
# Assign frequency_order flags
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['frequency_order'] >10) & (ords_prods_merge['frequency_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['frequency_order'] <=10, 'frequency_flag'] = 'Frequent customer'
```

```python
#checking counts
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Perform check on newest columns
ords_prods_merge[['user_id', 'frequency_order', 'frequency_flag']].head(20)
```
